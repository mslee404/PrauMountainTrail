```python
import requests
import time

GPX_FILE = "Gunung Prau.gpx"   # ganti sesuai file kamu
OUTPUT_CSV = "track_with_elevation.csv"

TARGET_TRACKS = [
    "via Wates 001",
    "via Dieng 001",
    "via Patakbanteng 001"
]

def get_elevation(lat, lon):
    """
    Query elevation from OpenElevation API.
    Format: https://api.open-elevation.com/api/v1/lookup?locations=lat,lon
    """
    url = f"https://api.open-elevation.com/api/v1/lookup?locations={lat},{lon}"

    try:
        r = requests.get(url, timeout=5)
        data = r.json()
        return data["results"][0]["elevation"]
    except:
        return None


def extract_points_from_gpx(gpx_file, target_tracks):
    gpx = gpxpy.parse(open(gpx_file, "r"))

    all_points = []

    for track in gpx.tracks:
        if track.name not in target_tracks:
            continue

        for segment in track.segments:
            for point in segment.points:
                all_points.append({
                    "track": track.name,
                    "latitude": point.latitude,
                    "longitude": point.longitude,
                })
    return pd.DataFrame(all_points)


def fetch_elevation_for_dataframe(df):
    elevations = []

    for idx, row in df.iterrows():
        lat = row["latitude"]
        lon = row["longitude"]

        elev = get_elevation(lat, lon)
        elevations.append(elev)

        print(f"[{idx+1}/{len(df)}] {lat}, {lon} -> {elev}")

        # Hindari rate limit
        time.sleep(0.2)

    df["elevation"] = elevations
    return df


# --- MAIN WORKFLOW ---
df_points = extract_points_from_gpx(GPX_FILE, TARGET_TRACKS)
df_with_elev = fetch_elevation_for_dataframe(df_points)
df_with_elev.to_csv(OUTPUT_CSV, index=False)

print(f"Done. Saved to {OUTPUT_CSV}")
```
